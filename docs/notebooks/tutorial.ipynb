## Running KNN normalization with an MuData object (using neighbors from RNA)

#### Write the protein and RNA data into a MuData object

```python
adata = sc.read_h5ad("/groups/nils/members/javier/Immune_colon/NB013_t001_filtered_protein_v1.h5ad")
rna = sc.read_h5ad("/groups/nils/members/javier/Immune_colon/NB013_t001_filtered_rna_v1.h5ad")
```

```python
mdata = mudata.MuData({"prot": adata, "rna": rna})
```

```python
mdata
```

#### Run KNN normalization on the MuData object, using calculate_neighbors_from="rna"

```python
knn.tl.knn_normalize(mdata, calculate_neighbors_from="rna", preprocess_rna=False)
```

The `preprocess_rna` argument indicates whether you want to library-normalize (sc.pp.normalize_total()) and log-transform (sc.pp.log1p()) the RNA data before calculating the KNN graph.

## Running KNN normalization on a pre-calculated KNN graph

Let's say you want to run KNN normalization but use your own KNN graph built with your own favorite method and data. This is only recommended if you think your KNN graph is more precise than our default KNN graph building procedure.

As an example, we run PCA on the protein data and use 20 PCs to calculate protein neighbors, using cosine distance.

```python
adata = sc.read_h5ad("/groups/nils/members/javier/Immune_colon/NB013_t001_filtered_protein_v1.h5ad")
```

```python
sc.pp.pca(adata, n_comps=20)
sc.pp.neighbors(adata, metric="cosine")
```

```python
adata
```

```python
knn.tl.knn_normalize(adata, calculate_neighbors_from="use_existing_neighbors")
```

The knn_normalize() function uses neighbors from adata.obsp["connectivities"].
